```python
import pandas as pd
import plotnine as p9

%matplotlib inline
```

```python
surveys_df = pd.read_csv("data/portal-teachingdb-master/surveys.csv")
```

```python
surveys_df.head()
```

```python
(p9.ggplot(data=surveys_df, 
           mapping = p9.aes(x="weight", y="hindfoot_length")) 
        + p9.geom_point())
```

```python
surveys_plot = (p9.ggplot(data=surveys_df, 
           mapping = p9.aes(x="weight", y="hindfoot_length")))
```

```python
surveys_plot + p9.geom_point()
```

```python
surveys_plot + p9.geom_bar()
```

## Using matplotlib with pandas

```python
surveys_df[["weight","hindfoot_length"]].plot(kind="scatter", x="weight",
                                             y="hindfoot_length")
```

```python
surveys_df["weight"].plot(kind="hist")
```

```python
surveys_df["weight"].plot(kind="kde")
```

```python
surveys_df[["weight","hindfoot_length"]].plot(kind="scatter", x="weight",
                                             y="hindfoot_length",
                                             color = "red",
                                             marker="o")
```

```python
import matplotlib.pyplot as plt

surveys_df[["weight","hindfoot_length"]].plot(kind="scatter", x="weight",
                                             y="hindfoot_length",
                                             color = "red",
                                             marker="o")

plt.title("My figure")
plt.xlabel("Weight (kg)")
plt.ylabel("Hindfoot length (cm)")
plt.savefig("figures/my-scatter.png", dpi=300)
```
