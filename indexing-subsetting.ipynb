## Workflow

```python
# repeate for every year
years = surveys_df["year"].unique()

for year in years:
    # select the year
    year_df = surveys_df[surveys_df["year"] == year]

    # create file name
    file_name = "yearly_files/surveys" + str(year) + ".csv"
    
    # save file
    year_df.to_csv(file_name)
```

```python
# collecting and concatenating dataframes

data_frame_collector = []

for file in os.listdir("yearly_files"):
    
    file_path = os.path.join("yearly_files",file)
    
    temp_file = pd.read_csv(file_path)
    
    data_frame_collector.append(temp_file)
    
# everything below here is outside the loop

big_df = pd.concat(data_frame_collector)

big_df = big_df.reset_index(drop=True)

big_df = big_df.drop("Unnamed: 0",axis=1)
```

```python
# functionize writing out to csv based on year

def one_year_csv(dataframe, output_path):
    
    assert dataframe.columns.isin(["year"]).any(), "Dataframe doesn't have a year column"
    
    # repeate for every year
    years = dataframe["year"].unique()

    for year in years:
        # select the year
        year_df = dataframe[dataframe["year"] == year]

        # create file name
        file_name = output_path +"/surveys" + str(year) + ".csv"
        # create_path(output_path, "surveys" + str(year),".csv")

        # save file
        year_df.to_csv(file_name)
```

```python
# function for creating file paths
def create_path(output_path, filename, file_ext):
    return os.path.join(output_path, filename + file_ext)
```

```python
one_year_csv(surveys_df, "yearly_files2")
```
